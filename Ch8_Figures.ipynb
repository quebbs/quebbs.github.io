```python
import matplotlib.pyplot as plt
import seaborn
%pylab inline
```

# Lagrange

```python
plt.figure(1,figsize=(5,3.5))
Q1=linspace(0,7)
Q2=linspace(0,7)
P1=25-(Q1-5.)**2
P2=16-(Q2-4.)**2
plt.plot(Q1,P1,ls='-',label='Unit-1')
plt.plot(Q2,P2,ls='-',label='Unit-2')
plt.xlabel('Unit Flow') ; plt.ylabel('Generation')
plt.legend(loc='upper left')
plt.ylim(0,26)
#Now, add second axis for total power generation as a fx of Q1
Q1=linspace(0,10)
Q2=10.-Q1
P1=25-(Q1-5.)**2
P2=16-(Q2-4.)**2
PT=P1+P2
plt.twinx()
plt.xlim(0,7)
plt.plot(Q1,PT,ls='--',c='r',label='Total Power f(Q1)')
plt.ylabel('Total Power $P_T$\n(as a function of Q1)')
plt.legend(loc='lower right')
plt.ylim(0,50)
plt.grid(False)
plt.tight_layout()
plt.savefig('./Figures/Lagrange_Power.png',dpi=140)
```

# Quadratic (Polynomial) Interpolation

```python
Y=[2,5,7,4.5,1]
X=[1,2,3,4,5]
plt.scatter(X,Y)
```

```python
#fourth order polynomial
from numpy import polyfit
polyfit(X,Y,len(X)-1)
```

```python
PF=polyfit(X,Y,len(X)-1)
x=linspace(0,6)
n=4
y=PF[0]*x**n+PF[1]*x**(n-1)+PF[2]*x**(n-2)+PF[3]*x**(n-3)+PF[4]*x**(n-4)
```

```python
plt.figure(1,figsize=(4,3))
plt.scatter(X,Y,s=70,c='r')
plt.plot(x,y)
plt.xlabel('Covariate \'X\'')
plt.ylabel('Response \'Y\'')
plt.tight_layout()
plt.savefig('./Figures/Ch8_Polynomial.png',dpi=140)
```
